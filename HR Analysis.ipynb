### Feature Engineering
- satisfaction_category (Low, Medium, High)
- tenure_bracket (0-2 years, 3-5 years, 6-10 years)
- Work_Life_Balance_Index (numeric ratio)
- ever_promoted (binary)
- salary_encoded (ordinal)

```python
# Define custom bins and labels
bins = [0.0, 0.4, 0.7, 1.0]   # Adjusting for a lower bound of 0.09
labels = ['Low', 'Medium', 'High']

# Create a new binned column
df['satisfaction_category'] = pd.cut(df['satisfaction_level'], bins=bins, labels=labels, include_lowest=True)

# Verify the transformation
print(df[['satisfaction_level', 'satisfaction_category']])
```

```python
# Define tenure bins and labels
tenure_bins = [0, 2, 5, 10]
tenure_labels = ['0-2 years', '3-5 years', '6-10 years']

# Create a tenure bracket column
df['tenure_bracket'] = pd.cut(df['time_spend_company'], bins=tenure_bins, labels=tenure_labels, right=True)

# Verify the new column
print(df[['time_spend_company', 'tenure_bracket']])
```

```python
# Calculate the Work-Life Balance Index (avoid division by zero)
df['Work_Life_Balance_Index'] = df['average_montly_hours'] / (df['number_project'] + 1)

# Optional: Scale for readability and round off
df['Work_Life_Balance_Index'] = df['Work_Life_Balance_Index'].round(2)

# Verify
print(df[['average_montly_hours', 'number_project', 'Work_Life_Balance_Index']])
```

```python
# Binary indicator: has the employee ever been promoted (1) or not (0)
df['ever_promoted'] = (df['promotion_last_5years'] > 0).astype(int)

# Verify
print(df[['promotion_last_5years', 'ever_promoted']])
```

```python
# Define salary order and map values
salary_order = {'low': 1, 'medium': 2, 'high': 3}
df['salary_encoded'] = df['salary'].map(salary_order)

# Verify
print(df[['salary', 'salary_encoded']])
```

```python
# Perform one-hot encoding
df = pd.get_dummies(df, columns=['Departments'], prefix='dept', drop_first=True)

# Verify the new columns
print(df.head())
```

Observations:

- Overlap: There is significant overlap between the two distributions, suggesting that employees with various levels of satisfaction may leave or stay.
- Peak Differences: The peak for employees who left is slightly lower than the peak for those who stayed, indicating a higher proportion of dissatisfied employees among those who left.

Inferences:
- Satisfaction Level and Turnover: Lower satisfaction levels are associated with a higher likelihood of turnover.
